```python
##ZOOM to one Company like BABA
import pandas as pd
from pandas_datareader import data
gDict={'谷歌':'GOOG','亚马逊':'AMZN','Facebook':'FB',
           '苹果':'AAPL','阿里巴巴':'BABA','腾讯':'0700.hk'}

start_date='2019-10-01'
end_date='2020-09-01'
babaDf=data.get_data_yahoo(gDict['阿里巴巴'],start_date,end_date)
def change(column):
    buyPrice=column[0]
    curPrice=column[233]
    priceChange=(curPrice-buyPrice)/buyPrice
    if(priceChange>0):
        print('股票累计上涨=',priceChange)
    elif(priceChange<0):
        print('股票累计无变化=',priceChange)
    else:
        print('股票累计下跌=',priceChange)
    return priceChange
#获取收盘价数据
closeCol=babaDf['Close']
#使用函数获取涨跌幅
babaChange=change(closeCol)
print(babaChange)
```
