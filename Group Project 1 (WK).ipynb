```python
url = r"https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
df_sp = pd.read_html(url)[0].reset_index()
df_sp.head()
```

```python
df_sp = pd.read_csv("sp500.csv")
df_sp.head(2)
```

```python
#Make this change in the base.py file in yfinance before running this code: https://github.com/ranaroussi/yfinance/pull/371/commits/7e137357296a1df177399d26543e889848efc021

for t in df_sp["Symbol"]:
    try:
        yfo = yf.Ticker(t)
        for k,v in yfo.info.items():
            if type(v) == list: 
                pass  
            else:
                df_sp.loc[df_sp["Symbol"] == t,k] = v
        print("{} completed".format(t))
    except:
        print("{} failed".format(t))
        continue
```

```python
df_sp.to_csv("sp500.csv",index=False)
```

```python
for c in sorted(df_sp.columns):
    print("{:<25}: {}".format(c,df_sp.loc[0,c]))
```

```python
sns.lmplot(x='marketCap', y='trailingPE', data=df_sp,
           fit_reg=False, 
           hue='sector')

for _, row in df_sp[(df_sp["trailingPE"]>2000) | (df_sp["marketCap"] > 0.7*10**12)].iterrows():
    plt.text(row["marketCap"], row["trailingPE"], row["Symbol"], horizontalalignment='left', size='medium', color='black', weight='normal')

plt.ylim(0, None) #300)
plt.xlim(0, None) #0.5 * 10 ** 12)
```

```python
# Set figure size with matplotlib
df1 = df_sp[df_sp["trailingPE"]<200]
plt.figure(figsize=(10,6))
plt.xticks(rotation=44)
 
# Create plot
sns.violinplot(x='sector',
               y='trailingPE', 
               data=df1, 
               inner=None) # Remove the bars inside the violins
               
sns.swarmplot(x='sector',
              y='trailingPE', 
              data=df1, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent


plt.title('PE by industry')
```

```python
sns.lmplot(x='heldPercentInsiders', y='trailingPE', data=df_sp,
           fit_reg=False, 
           hue='sector')

for _, row in df_sp[df_sp["heldPercentInsiders"]>0.3].iterrows():
    plt.text(row["heldPercentInsiders"], row["trailingPE"], row["Symbol"], horizontalalignment='left', size='medium', color='black', weight='normal')

plt.ylim(0, 300)
plt.xlim(0, None)
```

```python
# The mean return is the daily portfolio returns with the above four stocks.
# data_change = data.pct_change().dropna().mean(axis=1)

data_change = data["Adj Close"].pct_change().dropna()
data_change.plot()
plt.show()
```

```python
corr_df = data_change.corr(method='pearson')
corr_df
```

```python
mask = np.zeros_like(corr_df)
mask[np.triu_indices_from(mask)] = True

#generate plot

sns.heatmap(corr_df, cmap='RdYlGn', vmax=1.0, vmin=-1.0 , mask = mask, linewidths=2.5)
plt.yticks(rotation=0) 
plt.xticks(rotation=90) 
plt.show()
```
