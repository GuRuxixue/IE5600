```python
import pandas as pd 
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.gridspec as gridspec
```

```python
## Compare US with SG during cov-19
df_yahoo_A=yf.Ticker("^STI")
print(df_yahoo_A.info.keys())
df_yahoo_B=yf.Ticker("^DJI")
print(df_yahoo_B.info.keys())
history_A=df_yahoo_A.history(start="2019-09-01",end="2020-09-01")
history_B=df_yahoo_B.history(start="2019-09-01",end="2020-09-01")
#print(history_A)
fig, axs1 = plt.subplots(figsize=(16,9))
# plt.figure(figsize=(15,9))

# plt.subplot(211)

axs1.plot(history_A['Close'],label='SG')
#axs[0].set_title('SG')
axs1.set_xlabel("Date")
axs1.set_ylabel('Close',color='tab:blue')
axs1.grid(b=True,which='both',axis='both')
axs1.tick_params(axis='y', labelcolor='tab:blue')

fig.suptitle('SIT v.s.US 1-Year', fontsize=16)
axs2 = axs1.twinx() 
axs2.plot(history_B['Close'],color='tab:orange',label='US')
#axs[1].set_title('US')
axs2.set_xlabel("Date")
axs2.set_ylabel('Close',color='tab:orange')
axs2.tick_params(axis='y', labelcolor='tab:orange')
fig.legend(loc=1, bbox_to_anchor=(1,1), bbox_transform=axs1.transAxes)
# axs1.grid(b=True,axis='both')
```

```python
import matplotlib
import time
is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
    from IPython import display
df_yahoo_A=yf.Ticker("^STI")
# print(df_yahoo_A.info.keys())
df_yahoo_B=yf.Ticker("^DJI")
# print(df_yahoo_B.info.keys())
history_A=df_yahoo_A.history(start="2019-09-01",end="2020-09-01")
history_B=df_yahoo_B.history(start="2019-09-01",end="2020-09-01")
fig, axs1 = plt.subplots(figsize=(16, 9))
fig.suptitle('SG v.s.US 1-Year', fontsize=16)
axs2 = axs1.twinx()
e=[]
e1=[]
date = []
date2 = []
plt.ion()

axs1.set_xlabel("Date")
axs1.set_ylabel('Close', color='tab:blue')
axs1.grid(b=True, which='both', axis='both')
axs1.tick_params(axis='y', labelcolor='tab:blue')
axs1.set_xlim(history_A['Close'].index[0].date(), history_A['Close'].index[247].date())
axs1.set_ylim(2200, 3400)
axs2.set_ylabel('Close', color='tab:orange')
axs2.set_ylim(18000, 30000)
axs2.tick_params(axis='y', labelcolor='tab:orange')

for i in range(248):
    
    e.append(history_A['Close'][i])
    e1.append(history_B['Close'][i])
    date.append(history_A['Close'].index[i].date())
    date2.append(history_B['Close'].index[i].date())
    if i==100 and i%2==0:
        axs1.plot(date,e,label='SG',color='tab:blue',linestyle="-")
        axs2.plot(date2,e1, color='tab:orange', label='US',
                       linestyle="-")

        fig.legend(loc=1, bbox_to_anchor=(1, 1), bbox_transform=axs1.transAxes)
        if is_ipython:
            display.clear_output(wait=True)
            display.display(plt.gcf())
    elif i>100 and i%2==0:
        axs1.plot(date[i-2:i+1],e[i-2:i+1], color='tab:blue')
        axs2.plot(date2[i-2:i+1],e1[i-2:i+1], color='tab:orange')
        
        if i==246:
            axs1.annotate("",xy=(date[-1],e[-1]),xytext=(date[138],e[138]),arrowprops=dict(linewidth=3, mutation_scale=20,color='red',arrowstyle="->",connectionstyle="arc3"))
            axs2.annotate("",xy=(date2[-1],e1[-1]),xytext=(date2[136],e1[136]),arrowprops=dict(linewidth=3, mutation_scale=20,color='red',arrowstyle="->",connectionstyle="arc3"))
        if is_ipython:
            display.clear_output(wait=True)
            display.display(plt.gcf())
            
plt.close('all')
```
